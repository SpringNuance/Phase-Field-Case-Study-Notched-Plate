```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
H2_0_wtppm = np.loadtxt('H2_0_wtppm/FD_curve_0_wtppm.txt', skiprows=3)  
H2_0_wtppm = pd.DataFrame(H2_0_wtppm, columns=['time', 'displacement', 'force'])
H2_0p1_wtppm = np.loadtxt('H2_0p1_wtppm/FD_curve_0p1_wtppm.txt', skiprows=3)
H2_0p1_wtppm = pd.DataFrame(H2_0p1_wtppm, columns=['time', 'displacement', 'force'])
H2_0p5_wtppm = np.loadtxt('H2_0p5_wtppm/FD_curve_0p5_wtppm.txt', skiprows=3)
H2_0p5_wtppm = pd.DataFrame(H2_0p5_wtppm, columns=['time', 'displacement', 'force'])
H2_1_wtppm = np.loadtxt('H2_1_wtppm/FD_curve_1_wtppm.txt', skiprows=3)
H2_1_wtppm = pd.DataFrame(H2_1_wtppm, columns=['time', 'displacement', 'force'])

plt.figure(figsize=(7, 5))
plt.plot(H2_0_wtppm['displacement'], H2_0_wtppm['force'], label='0 wtppm (air)', linestyle='--')   
plt.plot(H2_0p1_wtppm['displacement'], H2_0p1_wtppm['force'], label='0.1 wtppm', linestyle='--')
plt.plot(H2_0p5_wtppm['displacement'], H2_0p5_wtppm['force'], label='0.5 wtppm', linestyle='--')
plt.plot(H2_1_wtppm['displacement'], H2_1_wtppm['force'], label='1 wtppm', linestyle='--')
plt.xlabel('Displacement (mm)')
plt.ylabel('Force (N)')
plt.xlim(-0.0001, 0.0101)
plt.ylim(-10, 800)
plt.legend(frameon=False)
```
